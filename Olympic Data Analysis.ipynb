# Importing Libraries:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Importing Dataset

```python
path = './Dataset/athlete_events.csv'
athlete = pd.read_csv(path)
athlete.head()
```

```python
path1 = './Dataset/noc_regions.csv'
regions = pd.read_csv(path1)
regions.head()
```

```python
# Shape of both the dataframes

print("The shape of Athlete dataframe is:", athlete.shape)
print("The shape of Regions dataframe is:", regions.shape)
```

```python
# Merging the two dataset

athlete_df = athlete.merge(regions, how='left', on='NOC')
athlete_df.head()
```

```python
# Shape of combined dataframe

row, column = athlete_df.shape
print("The number of rows are:", row)
print("The number of columns are:", column)
```

```python
# Renaming column

athlete_df.rename(columns={'region':'Region', 'notes':'Notes'}, inplace=True)
athlete_df.head()
```

```python
# Check for null values

athlete_df.isnull().sum()
```

```python
# Info about the dataset

athlete_df.info()
```

```python
# Statistical information about the numerical fields

athlete_df.describe()
```

```python
# India Details

india = athlete_df[athlete_df['Team'] == 'India']
india.head()
```

```python
# Japan details

japan = athlete_df[athlete_df['Team'] == 'Japan']
japan.head()
```

```python
# Top 10 participating countries

top_10 = athlete_df['Team'].value_counts().sort_values(ascending = False).head(10)
top_10
```

```python
plt.figure(figsize=(15, 9))
sns.barplot(x = top_10.index, y = top_10, palette='Set2' )
plt.title("Top 10 Countries")
plt.show()
```

```python
# Age distribution 

plt.figure(figsize=(15, 9))
plt.hist(athlete_df['Age'], bins= np.arange(athlete_df['Age'].min(), athlete_df['Age'].max(), 2), color='orange', edgecolor = 'white')
plt.title("Age Distribution of Athletes")
plt.xlabel("Age")
plt.xticks(np.arange(athlete_df['Age'].min(), athlete_df['Age'].max(), 5))
plt.ylabel("Number of Participants")
plt.show()
```

```python
# Sports held at winter olympics

winter_sports = athlete_df[athlete_df['Season'] == 'Winter']['Sport'].unique()
winter_sports
```

```python
# Sports held at Summer Olympics

summer_sports = athlete_df[athlete_df['Season'] == 'Summer']['Sport'].unique()
summer_sports
```

```python
# Male and Female participants

gender = athlete_df['Sex'].value_counts()
male = gender[0]
female = gender[1]
print("Number of males are:", male)
print("Number of females are:", female)
```

```python
plt.figure(figsize=(12, 9))
plt.pie(x = [male, female], labels=["Male", "Female"], autopct='%1.1f%%', startangle=150)
plt.show()
```

```python
# Total Medals

athlete_df['Medal'].value_counts().to_frame()
```

```python
# Total number of female athletes in each olympic (Summer)

only_female = athlete_df[(athlete_df['Sex'] == 'F') & (athlete_df['Season'] == 'Summer')][['Sex', 'Year']]
only_female = only_female.groupby('Year').count().reset_index()
only_female.tail()
```

```python
# Renaming the column

only_female.rename(columns={'Sex': 'Count'},  inplace=True)
only_female.tail() # To see the latest trend
```

```python
plt.figure(figsize=(15, 9))
sns.barplot(x = 'Year', y = 'Count', data = only_female)
plt.title("Women Participation")
plt.show()
```

```python
plt.figure(figsize=(15, 9))
plt.plot(only_female['Year'], only_female['Count'])
plt.xlabel("Year")
plt.ylabel("Count")
plt.title("Women Participation")
plt.show()
```

```python
# Gold medal Athletes

goldMedal = athlete_df[athlete_df['Medal'] == 'Gold']
goldMedal
```

```python
# Gold above more than 60 age

goldMedal[goldMedal['Age'] > 60]
```

```python
# Sports for which these gold medal has come

sports = goldMedal['Sport'][goldMedal['Age'] > 60 ]
sports
```

```python
#Plotting the above result
plt.figure(figsize=(12, 9))
sns.countplot(sports)
plt.title("Gold Medal for Athletes over the Age of 60 ")
plt.show()
```

```python
# Gold Medals for each country

goldMedal_country = goldMedal['Region'].value_counts().reset_index(name = 'Medals').head(6)
goldMedal_country.rename(columns={'index':'Region'}, inplace=True)
goldMedal_country
```

```python
plt.figure(figsize=(12, 9))
sns.barplot(x = 'Region', y = 'Medals', data = goldMedal_country, palette='crest')
plt.title("Top 6 Gold Medal Winning Countries")
plt.show()
```

```python
# Rio olympics 

rio_year = athlete_df['Year'].max()

team_rio = athlete_df[(athlete_df['Year'] == 2016) & (athlete_df['Medal'] == 'Gold')]['Team']
top_20_rio = team_rio.value_counts().head(20)
top_20_rio
```

```python
plt.figure(figsize=(15, 9))
sns.barplot(x = top_20_rio, y = top_20_rio.index)
plt.title("Top 20 Gold Medal winning Teams")
plt.xlabel("Number of Medals")
plt.ylabel("Teams")
plt.show()
```

```python
# Filter out only athlete who have won medals

df_medal = athlete_df.dropna(subset=['Medal'], axis = 0).reset_index()
df_medal
```

```python
# Height vs Weight scatter plot for athletes who won medal 

df_medal_HW = df_medal[(df_medal['Height'].notnull()) & (df_medal['Weight'].notnull())]
```

```python
plt.figure(figsize=(15, 9))
sns.scatterplot(x = 'Height', y = 'Weight', data = df_medal_HW, hue='Sex')

plt.show()
```
